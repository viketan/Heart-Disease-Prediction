## 1) Problem statement
This data set dates from 1988 and consists of four databases: Cleveland, Hungary, Switzerland, and Long Beach V. It contains 76 attributes, including the predicted attribute, but all published experiments refer to using a subset of 14 of them. The "target" field refers to the presence of heart disease in the patient. It is integer valued 0 = no disease and 1 = disease.

## 2) Data Collection.
- The Dataset is collected from https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset
- The data consists of 14 column and 1025 rows.

## 3) Attribute infomation
- age
- sex: sex (1 = male; 0 = female)
- cp: chest pain type (4 values)
    - Value 0: asymptomatic
    - Value 1: typical angina
    - Value 2: atypical angina
    - Value 3: non-anginal pain

- trestbps: resting blood pressure (in mm Hg on admission to the hospital)
- chol: serum cholestoral in mg/dl
- fbs: fasting blood sugar > 120 mg/dl
- restegc: resting electrocardiographic results
    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
- thalach: maximum heart rate achieved
- exang: exercise induced angina (1 = yes; 0 = no)
- oldpeak = ST depression induced by exercise relative to rest
- slope: the slope of the peak exercise ST segment
    - Value 1: upsloping
    - Value 2: flat
    - Value 3: downsloping
- ca : number of major vessels (0-3) colored by flourosopy
- thal: 0 = normal; 1 = fixed defect; 2 = reversable defect
- target: diagnosis of heart disease (angiographic disease status)
    - Value 0: < 50% diameter narrowing
    - Value 1: > 50% diameter narrowing

```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
categorical_features.remove('target')
# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

## Analysis of Categorical features

```python
for col in categorical_features:
    print(df[col].value_counts(normalize=True) * 100)
    print('---------------------------')
```

```python
plt.figure(figsize=(30, 30))
plt.suptitle('Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
for i in range(0, len(categorical_features)):
    plt.subplot(5, 3, i+1)
    img=sns.countplot(x=categorical_features[i],data=df,hue='target')
    plt.xticks(rotation=90);
    for container in img.containers:
        img.bar_label(container);
    plt.xlabel(categorical_features[i])
    plt.tight_layout()
```

```python
def analysis(col):
    count = df.groupby(col)["target"].value_counts(normalize=True)
    count_df = pd.DataFrame(count)
    count_df.rename(columns = {"target" : "percentage"}, inplace = True)
    count_df.reset_index(inplace = True)
    count_df.sort_values(by = ["target", col], inplace=True)
    plt.figure(figsize = (10, 10))
    index = 1
    for c in list(count_df[col].unique()):
        plt.subplot(1,len(list(count_df[col].unique())),index)
        count_df.groupby(col)["percentage"].get_group(c).plot.pie(subplots=True,
                                                 labels=list(count_df['target'].unique()),
                                                 autopct="%.2f%%",
                                                 textprops={'fontsize': 10},
                                                 )
        plt.title(c, fontdict = {'fontsize': 14})
        index += 1
```

```python
for col in categorical_features:
    analysis(col)
```

#### Insights

- comapared to male, females are prone to heart disease.
- If you have chest pain, most likely you are suffering from heart disease.
- There is no correlation between blood sugar level and heart disease.
- abnormal rest ecg have more chance of heart disease.
- If you have chest pain while doing any phyisical activity, it may beacuse of underlying heart diseases.
